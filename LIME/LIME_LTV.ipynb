```python
import pandas as pd
```

```python
loans = pd.read_excel('ltv1.xlsx')
```

```python
loans[loans['is_insurance_paid'] == 1].head(20)
```

```python
def get_real_loan_value(row):
    if row['is_insurance_paid'] == 1 or row['is_insurance_paid'] == 2:
        return row['value'] / 111 * 100
    else:
        return row['value']
```

```python
loans['requested_value'] = loans.apply(get_real_loan_value, axis=1)
loans['insurance'] = loans['value'] - loans['requested_value']
loans = loans[['cid', 'timestamp', 'status', 'is_insurance_paid', 
               'requested_value', 'pct', 'insurance', 'commis', 'fees']]
```
